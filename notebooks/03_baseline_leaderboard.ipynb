# Baseline Leaderboard ViewerLoad the leaderboard JSON generated by the CLI and display it as tables.Highlight best-null-por-CP and the regime segments.Purely exploratory — not load-bearing.

```python
import jsonfrom pathlib import Pathimport polars as pl# Point this at the most recent leaderboard JSON produced by the CLI.lb_dir = Path('./reports/leaderboard')candidates = sorted(lb_dir.glob('*-leaderboard.json'))board = json.loads(candidates[-1].read_text())print('window:', board['window'])print('best nulls:', board.get('best_nulls', {}))
```

```python
# Per-CP tablefor cp, entries in board.get('by_cp', {}).items():    print(f'== CP {cp} (best null: {board["best_nulls"].get(cp)}) ==')    display(pl.DataFrame(entries).sort('mae'))
```

```python
# Segment tablefor seg_name, seg_entries in board.get('segments', {}).items():    print(f'== segment: {seg_name} ==')    display(pl.DataFrame(seg_entries).sort('mae'))
```
